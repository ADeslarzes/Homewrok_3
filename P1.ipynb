### $\textbf{Exercise 1 :}$

#### $\textbf{Q1 :}$

##### $\textbf{a) :}$

Nous considérons un système de $N$ particules avec l'hamiltonien suivant :
$$ H_N(\vec x_1, ..., \vec x_N; \vec p_1, ..., \vec p_N)= \sum_{i=1}^N\frac{\| \vec p_i \|^2_2}{2m}+\sum_{i<j}u(r_{ij}), \qquad \text{avec } r_{ij}=\|\vec x_i-\vec x_j \|_2 $$
Où $\vec x_i$ et $\vec p_i$ sont la postion set la quantité de mouvement de la particule $i$.$\\$
Alors la fonction de partition $Z_N$ s'écrit :
$$ Z_N=\frac{1}{N! h^{3N}}\int e^{-\beta (\sum_{i=1}^N\frac{\| \vec p_i \|^2_2}{2m}+\sum_{i<j}u(r_{ij}))}\text{d}^{3N}p \text{ d}^{3N}r= \frac{1}{N! h^{3N}}\prod_{i=1}^N\int \text{d}p_{xi} e^{-\beta \frac{p_{xi}^2}{2m}}\int \text{d}p_{yi} e^{-\beta \frac{p_{yi}^2}{2m}}\int \text{d}p_{zi} e^{-\beta \frac{p_{zi}^2}{2m} }\int\text{d}^{3N}r e^{-\beta\sum_{i<j}u(r_{ij})} \\ \implies Z_N=\frac{1}{N! h^{3N}} (\sqrt{2\pi m k_B T})^{3N}\int\text{d}^{3N}r e^{-\beta\sum_{i<j}u(r_{ij})}=\frac{1}{N! h^{3N}\lambda^{3N}} \int\text{d}^{3N}r e^{-\beta\sum_{i<j}u(r_{ij})}$$
Ceci implique donc :
$$ Q_N(V,T)=\int\text{d}^{3N}r e^{-\beta\sum_{i<j}u(r_{ij})} $$

##### $\textbf{b) :}$

En considérant acune intéraction entre les particules, donc en posant $u=0$, nous obtenons :
$$ Z_N=\frac{1}{N! \lambda^{3N}}\int \text{d}^{3N}r 1=\frac{V^N}{N! \lambda^{3N}} $$

#### $\textbf{Q2 :}$

Posons $u_{ij}=u(r_{ij})$ et $f_{ij}=e^{-\beta u_{ij}}-1$.

##### $\textbf{a) :}$

$f_{ij}$ représente a quelle point l'interaction entre $\vec x_i$ et $\vec x_j$ est forte ainsi que le type d'intéraction. Pour voir cela intéressons nous a plusieurs valeur de $f_{ij}$. $\\$
Considérons sans intéractions : 
$$ u_{ij}=0 \implies f_{ij}=0$$
Si nous considérons une forte répulsion, donc $u_{ij}>> k_B T$, alors :
$$ e^{-\beta u_{ij}} \approx 0\implies f_{ij}\approx -1 $$
Et si nous considérons une forte attraction donc $u_{ij}<< k_B T$, alors :
$$ e^{\beta u_{ij}} >> 1 \implies f_{ij}\approx e^{-\beta u_{ij}} $$
Comme $e^x$ est strictement croissant, il est directement observable que le signe de $f_{ij}$ dicte le type de force. Si $f_{ij}\leq 0$ alors il y a une répulsion. Si $f_{ij}\geq 0$ alors il y a une attraction. Plus $f_{ij}$ est garnd plus l'attraction est forte et plus $f_{ij}$ est proche de 1 plus la répulsion est forte. 

##### $\textbf{b) :}$

Réécrivons l'intégrale de configuration $Q_N(V,T)$ en posant $u(r_{ij})=u_{ij}$ : $\\$
Par définition de $f_{ij}$ nous avons :
$$ e^{-\beta u_{ij}} = 1+f_{ij} \implies Q_N(V,T)=\int e^{-\beta \sum_{i<j}u_{ij}} \text{d}^{3N}r=\int\prod_{i<j} e^{-\beta u_{ij}} \text{d}^{3N}r=\int\prod_{i<j} (1+f_{ij}) \text{d}^{3N}r$$

#### $\textbf{Q3 :}$

##### $\textbf{a) :}$

Pour $N=2$ particules, considérons l'intégrale de configuration :
$$ Q_2=\int \int e^{-\beta u_{12}} \text{d}^3r_1\text{d}^3r_2 $$
En se rappelant que $f_{ij}:=e^{-\beta u_{ij}}-1$ il est possible de réécrire $Q_2$ comme :
$$ Q_2=\int \int(1+f_{ij})\text{d}^3r_1\text{d}^3r_2=V^2+\int \int f(\| \vec r_1 -\vec r_2 \|) \text{d}^3r_1\text{d}^3r_2= V^2 + \int \text{d}^3R \int \text{d}^3r_{12} f(\| \vec r_{12} \|)=V^2 + \int \text{d}^3R \int \text{d}^3r_{12} f_{12} $$
Où $\vec R=\frac{1}{2}(\vec r_1 + \vec r_2)$ et $vec r_{12}= \vec r_1 -\vec r_2$. Ce qui donne :
$$
\vec R=\begin{pmatrix}
X \\
Y \\
Z
\end{pmatrix}, \qquad 
\vec r_{12}=\begin{pmatrix}
r_1 \\
r_2 \\
r_3
\end{pmatrix}
$$
Avec $X=\frac{1}{2}(x_1+x_2)$, $Y=\frac{1}{2}(y_1+y_2)$, $Z=\frac{1}{2}(z_1+z_2)$ et $r_1=x_1-x_2$, $r_2=y_1-y_2$ et $r_3=z_1-z_2$. $\\$
Dans ce cas le jacobine $J=1$ 

##### $\textbf{b) :}$

Considérons maintenant 3 particules. $\\$
Rappelons d'abord la définition de $Q_3$ : 
$$ Q_3=\int \int \int \prod_{i<j} (1+f_{ij}) \text{d}^9r, \qquad i,j\in \{ 1, 2,3\} $$
Intéressons nous d'abord à l'intégrande :$\\$
$$ \prod_{i<j}(1+f_{ij})=(1+f_{12})(1+f_{13})(1+f_{23})=(1+f_{12}+f{13}+f_{12}f_{13})(1+f_{23})=1+f_{12}+f_{13}+f_{23}+f_{12}f_{13}+f_{12}f_{23}+f_{13}f_{23}+f_{12}f_{13}f_{23} $$
Ceci nous permet donc d'écire $Q_3$ comme :
$$ \int \int \int (1+f_{12}+f_{13}+f_{23}+f_{12}f_{13}+f_{12}f_{23}+f_{13}f_{23}+f_{12}f_{13}f_{23})\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 $$

##### $\textbf{c) :}$

En utilisant la linéarité de l'intégrale : 
$$ Q_3=\int \int \int 1 \text{d}^3r_1\text{d}^3r_2\text{d}^3r_3+ \int \int \int f_{12}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 + \int \int \int f_{13}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 + \int \int \int f_{23}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 + \int \int \int f_{12}f_{13}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 \\ + \int \int \int f_{12}f_{23}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 + \int \int \int f_{13}f_{23} \text{d}^3r_1\text{d}^3r_2\text{d}^3r_3+ \int \int \int f_{12}f_{13}f_{23}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 \\ = V^3 + V \int \int f_{12}\text{d}^3r_1\text{d}^3r_2 + V \int \int f_{13}\text{d}^3r_1\text{d}^3r_3 + V \int \int f_{23}\text{d}^3r_2\text{d}^3r_3 + \int \int \int f_{12}f_{13}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 \\ + \int \int \int f_{12}f_{23}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 + \int \int \int f_{13}f_{23} \text{d}^3r_1\text{d}^3r_2\text{d}^3r_3+ \int \int \int f_{12}f_{13}f_{23}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 $$
En remarquer que les variables d'intégration sont des variables muettes : 
$$ Q_3= V^3 + 3V \int \int f_{12}\text{d}^3r_1\text{d}^3r_2  + \int \int \int f_{12}f_{13}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 \\ + \int \int \int f_{12}f_{23}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 + \int \int \int f_{13}f_{23} \text{d}^3r_1\text{d}^3r_2\text{d}^3r_3+ \int \int \int f_{12}f_{13}f_{23}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 $$
En posant le changement de variable pour l'intégrale avec $f_{ij}f_{ik}$, $\vec r_{ij} =\vec r_i-vec r_j$, $\vec r_{ik}=\vec r_i-\vec r_k$ et $\vec R= \frac{1}{3}(\vec r_1+\vec r_2+\vec r_3)$ et en évaluant le jacobien de la transformation, $J=1$. Il est possible de remarquer que les trois intégrales sur les paires $f_{ij}f_{ik}$ sont indentiques (à variables muettes près) et que l'intégrande est indépendante de $\vec R$ (comme à la question 2). Ce qui donne :
$$Q_3=Q_3= V^3 + 3V \int \int f_{12}\text{d}^3r_1\text{d}^3r_2  + 3V \int \int f_{12}f_{13}\text{d}^3r_{12}\text{d}^3r_{13} + \int \int \int f_{12}f_{13}f_{23}\text{d}^3r_1\text{d}^3r_2\text{d}^3r_3 $$
En appliquant le même changement de variable à la dernière intégrale triple, et en remarquant que $\vec r_{23}= \vec r_{13}-\vec r_{12}$, nous en concluons que l'intégrande est indépendante de $\vec R$ et donc :
$$Q_3=Q_3= V^3 + 3V \int \int f_{12}\text{d}^3r_1\text{d}^3r_2  + 3V \int \int f_{12}f_{13}\text{d}^3r_{12}\text{d}^3r_{13} + V\int \int f_{12}f_{13}f_{23}\text{d}^3r_{12}\text{d}^3r_{13} $$

#### $\textbf{Q4 :}$

##### $\textbf{b) :}$

Montrons la relations (19). $\\$
Rappelons que :
$$ J=-\frac{1}{\beta}\log(\Theta)=-\frac{1}{\beta}\left( \sum_{l=1}^\infty b_lz^l\frac{V}{\lambda^3}\right) $$
Ainsi que :
$$ P=-\frac{\partial J}{\partial V}=\frac{1}{\beta}\sum_{l=1}^\infty b_l\frac{z^l}{\lambda^3}\implies \frac{p}{k_B T}=\sum_{l=1}^\infty b_l\frac{z^l}{\lambda^3} $$
Ce qui est la relation souhaitée.

##### $\textbf{c) :}$

Montrons maintenant la relation (20). $\\$
Rappelons que :
$$ N=-\frac{\partial J}{\partial \mu},\quad \text{et} \quad \frac{\partial z}{\partial \mu}=\beta z $$
Ceci nous donnes donc :
$$ N=\frac{1}{\beta}\sum_{l=1}^\infty b_l l z^{l-1}\beta z \frac{V}{\lambda^3}=\frac{V}{\lambda^3}\sum_{l=1}^\infty b_l l z^l\implies \frac{N}{V}=\frac{1}{\lambda^3}\sum_{l=1}^\infty b_l l z^l $$
Ce qui est la relation souhaitée
