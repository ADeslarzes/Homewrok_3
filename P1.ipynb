### $\textbf{Exercise 1 :}$

#### $\textbf{Q1 :}$

##### $\textbf{a) :}$

Nous considérons un système de $N$ particules avec l'hamiltonien suivant :
$$ H_N(\vec x_1, ..., \vec x_N; \vec p_1, ..., \vec p_N)= \sum_{i=1}^N\frac{\| \vec p_i \|^2_2}{2m}+\sum_{i<j}u(r_{ij}), \qquad \text{avec } r_{ij}=\|\vec x_i-\vec x_j \|_2 $$
Où $\vec x_i$ et $\vec p_i$ sont la postion set la quantité de mouvement de la particule $i$.$\\$
Alors la fonction de partition $Z_N$ s'écrit :
$$ Z_N=\frac{1}{N! h^{3N}}\int e^{-\beta (\sum_{i=1}^N\frac{\| \vec p_i \|^2_2}{2m}+\sum_{i<j}u(r_{ij}))}\text{d}^{3N}p \text{ d}^{3N}r= \frac{1}{N! h^{3N}}\prod_{i=1}^N\int \text{d}p_{xi} e^{-\beta \frac{p_{xi}^2}{2m}}\int \text{d}p_{yi} e^{-\beta \frac{p_{yi}^2}{2m}}\int \text{d}p_{zi} e^{-\beta \frac{p_{zi}^2}{2m} }\int\text{d}^{3N}r e^{-\beta\sum_{i<j}u(r_{ij})} \\ \implies Z_N=\frac{1}{N! h^{3N}} (\sqrt{2\pi m k_B T})^{3N}\int\text{d}^{3N}r e^{-\beta\sum_{i<j}u(r_{ij})}=\frac{1}{N! h^{3N}\lambda^{3N}} \int\text{d}^{3N}r e^{-\beta\sum_{i<j}u(r_{ij})}$$
Ceci implique donc :
$$ Q_N(V,T)=\int\text{d}^{3N}r e^{-\beta\sum_{i<j}u(r_{ij})} $$

##### $\textbf{b) :}$

En considérant acune intéraction entre les particules, donc en posant $u=0$, nous obtenons :
$$ Z_N=\frac{1}{N! \lambda^{3N}}\int \text{d}^{3N}r 1=\frac{V^N}{N! \lambda^{3N}} $$

#### $\textbf{Q2 :}$

Posons $u_{ij}=u(r_{ij})$ et $f_{ij}=e^{-\beta u_{ij}}-1$.

##### $\textbf{a) :}$

$f_{ij}$ représente a quelle point l'interaction entre $\vec x_i$ et $\vec x_j$ est forte ainsi que le type d'intéraction. Pour voir cela intéressons nous a plusieurs valeur de $f_{ij}$. $\\$
Considérons sans intéractions : 
$$ u_{ij}=0 \implies f_{ij}=0$$
Si nous considérons une forte répulsion, donc $u_{ij}>> k_B T$, alors :
$$ e^{-\beta u_{ij}} \approx 0\implies f_{ij}\approx -1 $$
Et si nous considérons une forte attraction donc $u_{ij}<< k_B T$, alors :
$$ e^{\beta u_{ij}} >> 1 \implies f_{ij}\approx e^{-\beta u_{ij}} $$
Comme $e^x$ est strictement croissant, il est directement observable que le signe de $f_{ij}$ dicte le type de force. Si $f_{ij}\leq 0$ alors il y a une répulsion. Si $f_{ij}\geq 0$ alors il y a une attraction. Plus $f_{ij}$ est garnd plus l'attraction est forte et plus $f_{ij}$ est proche de 1 plus la répulsion est forte. 

##### $\textbf{b) :}$

Réécrivons l'intégrale de configuration $Q_N(V,T)$ en posant $u(r_{ij})=u_{ij}$ : $\\$
Par définition de $f_{ij}$ nous avons :
$$ e^{-\beta u_{ij}} = 1+f_{ij} \implies Q_N(V,T)=\int e^{-\beta \sum_{i<j}u_{ij}} \text{d}^{3N}r=\int\prod_{i<j} e^{-\beta u_{ij}} \text{d}^{3N}r=\int\prod_{i<j} (1+f_{ij}) \text{d}^{3N}r$$

#### $\textbf{Q4 :}$

##### $\textbf{a) :}$

En prenant l'expression donnée pour $Z_N$ à savoir 
$$
Z_N = \sum_{\{m_l\}}\prod_{l=1}^N\frac{(b_l V/\lambda^3)^{m_l}}{m_l!}
$$
Nous pouvons la remplacer dans $\Xi = \sum_{N=0}^\infty z^N Z_N$. Cela donne
$$
\Xi = \sum_{N=0}^\infty z^N \sum_{\{m_l\}}\prod_{l=1}^N\frac{(b_l V/\lambda^3)^{m_l}}{m_l!}
$$
En utilisant la condition $N = \sum_{l=1}^Nlm_l$, nous pouvons remplacer l'exposant de $z$ et exprimer $z^N = z^{\sum_{l=1}^Nlm_l}$ à travers le produit de telle manière à avoir 
$$
\Xi = \sum_{N=0}^\infty \sum_{\{m_l\}}\prod_{l=1}^N\frac{(b_l z^l V/\lambda^3)^{m_l}}{m_l!}
$$

Si l'on regarde notre fonction, nous prenons la somme sur toutes les configurations $\{m_l\}$ telles que $N = \sum_{l=1}^Nlm_l$ et ensuite une somme sur toutes les valeurs possibles de $N$. Cette double somme peut être réécrite comme une somme sur toutes les configurations $\{m_l\}$ possibles. En faisant cela, il faut également changer la borne du produit et prendre le produit infini sur les valeurs de $l$. En effet pour une configuration donnée, prendre ce produit infini ne va garder que les termes où $m_l \neq 0$. Les termes où $m_l = 0$ ne vont contribuer que par un facteur 1 ce qui ne change pas le produit. Ainsi on obtient l'expression suivante :
$$
\Xi = \sum_{\{m_l\}}\prod_{l=1}^\infty \frac{1}{m_l!} \left(\frac{b_lz^lV}{\lambda^3}\right)^{m_l}
$$

Afin d'arriver au résultat désiré il ne reste qu'à remarquer que prendre la somme sur toutes les configurations $\{m_l\}$ du produit sur $l$ est strictement équivalent à prendre le produit sur $l$ de la somme sur toutes les valeurs possible de $m_l$. C'est à dire : 
$$
\Xi = \sum_{\{m_l\}}\prod_{l=1}^\infty \frac{1}{m_l!} \left(\frac{b_lz^lV}{\lambda^3}\right)^{m_l} = \prod_{l=1}^\infty \sum_{m_l=0} ^\infty \frac{1}{m_l!} \left(\frac{b_lz^lV}{\lambda^3}\right)^{m_l}
$$
On reconnait la définition de l'exponentielle. Finalement, le produit d'exponentielle et l'exponentielle de la somme, donnant ainsi
$$
\Xi =  \prod_{l=1}^\infty \sum_{m_l=0} ^\infty \frac{1}{m_l!} \left(\frac{b_lz^lV}{\lambda^3}\right)^{m_l} =  \prod_{l=1}^\infty \exp{\left(\frac{b_lz^lV}{\lambda^3}\right)} = \exp{\left(\sum_{l=1}^\infty \frac{b_lz^lV}{\lambda^3}\right)}
$$

##### $\textbf{d) :}$

Prenons les résultats précédants : 
$$
\frac{P}{k_bT} = \frac{1}{\lambda^3}\sum_{l=1}^\infty b_l z^l \quad (1) \quad \text{et}\quad \frac{N}{V} = \frac{1}{\lambda^3}\sum_{l=1}^\infty l b_lz^l := \frac{y}{\lambda^3} \quad (2)
$$
Prenons $\frac{\partial y}{\partial z}\bigr\rvert_{z = 0}$ :
$$
\frac{\partial y}{\partial z}\bigr\rvert_{z = 0} = \sum_{l=1}^\infty l^2 b_lz^{l-1}\bigr\rvert_{z = 0} = b_1 \neq 0
$$

Ainsi par le théorème d'inversion locale il existe une fonction analytique $f$ telle que $z = f(y)$ dans un voisinage de $y=0$. Prenons le développement de Taylor de $f$:
$$
z = f(y) = \sum_{n=0}^\infty c_n y^n
$$
Or nous avons que $y(0) = 0$ et comme $f$ constitue une bijection sur le voisinage, nécessairement $a_0 = 0$. Ainsi 
$$
f(y) = \sum_{n=1}^\infty c_n y^n
$$
En injectant ce résultat dans l'équation $(1)$ nous obtenons 
$$
\frac{P}{k_bT} = \frac{1}{\lambda^3}\sum_{l=1}^\infty b_l f(y)^l
$$
Comme $f(y)$ est une série de puissance, $f(y)^l$ est à nouveau une série de puissance et sa somme infinie est encore une série de puissance. Finalement on peut écrire 
$$
\frac{P}{k_bT} = \frac{1}{\lambda^3}\sum_{l=1}^\infty a_l y^l = \frac{1}{\lambda^3}\sum_{l=1}^\infty a_l\left(\frac{\lambda^3 N}{V}\right)^l \iff \frac{PV}{Nk_bT} = \sum_{l=1}^\infty a_l \left(\frac{N\lambda^3}{V}\right)^{l-1} = \sum_{l=1}^\infty a_l \left(\frac{\lambda^3}{v}\right)^{l-1}
$$
Ce qui démontre l'équation du viriel.
